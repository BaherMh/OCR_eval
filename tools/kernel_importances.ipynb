```python
import numpy as np

# ----------------------------
# Compute per-filter "peak after batch-mean" activation
# ----------------------------
def compute_per_filter_peak_activation(outputs_dict):
    """
    For each layer in outputs_dict:
        output shape: [B, C, H, W]
        -> for each filter c:
            abs_out = |output[:, c, :, :]|
            mean_over_batch = mean(abs_out, axis=0)  # [H, W]
            max_value = max(mean_over_batch)         # scalar
        returns: dict {layer_name: np.array of shape [C]}
    """
    result = {}
    for name, out in outputs_dict.items():
        # out: [B, C, H, W]
        abs_out = paddle.abs(out)  # [B, C, H, W]
        
        # Mean over batch dimension (axis=0) -> [C, H, W]
        mean_over_batch = paddle.mean(abs_out, axis=0)  # [C, H, W]
        
        # Max over spatial dims (H, W) -> [C]
        max_per_filter = paddle.max(mean_over_batch, axis=[1, 2])  # [C]
        
        result[name] = max_per_filter.numpy()  # shape: (C,)
    return result

# Compute for both models
activations_base = compute_per_filter_peak_activation(conv_outputs_base)
activations_tuned = compute_per_filter_peak_activation(conv_outputs_tuned)

# ----------------------------
# Compute ratios (tuned / base) with safe division
# ----------------------------
activation_ratios = {}
common_layers = set(activations_base.keys()) & set(activations_tuned.keys())
print(f"\nComparing {len(common_layers)} common Conv2D layers\n")

epsilon = 1e-8
for name in sorted(common_layers):
    base_vals = activations_base[name]      # [C]
    tuned_vals = activations_tuned[name]    # [C]
    
    # Ratio per filter
    ratio = tuned_vals / (base_vals + epsilon)
    activation_ratios[name] = ratio
    
    # Stats
    max_ratio = np.max(ratio)
    min_ratio = np.min(ratio)
    mean_ratio = np.mean(ratio)
    max_idx = np.argmax(ratio)
    
    print(f"Layer: {name}")
    print(f"  Num filters: {len(ratio)}")
    print(f"  Ratio per filter → min: {min_ratio:.3f}, mean: {mean_ratio:.3f}, max: {max_ratio:.3f}")
    print(f"  Filter with highest ratio: {max_idx} (ratio = {max_ratio:.3f})")
```

```python
# ----------------------------
# Create ternary indicator per filter: -1, 0, 1
# ----------------------------
filter_change_indicators = {}  # {layer_name: list of -1, 0, 1 (length = num_filters)}

for name, ratio in activation_ratios.items():
    ratio = np.array(ratio)  # ensure numpy array
    
    # Initialize indicator array with zeros
    indicator = np.zeros_like(ratio, dtype=int)
    
    # Set 1 where ratio > 1.1
    indicator[ratio > 1.1] = 1
    
    # Set -1 where ratio < 0.9
    indicator[ratio < 0.9] = -1
    
    filter_change_indicators[name] = indicator.tolist()  # convert to Python list

    # Optional: print summary
    num_filters = len(indicator)
    num_increased = int(np.sum(indicator == 1))
    num_decreased = int(np.sum(indicator == -1))
    num_unchanged = num_filters - num_increased - num_decreased
    print(f"Layer: {name} → ↑{num_increased} / ↓{num_decreased} / ~{num_unchanged} filters")

print("\nExample for first layer:")
first_layer = next(iter(filter_change_indicators))
print(f"Layer '{first_layer}': {filter_change_indicators[first_layer][:15]}...")  # first 15 values
```
